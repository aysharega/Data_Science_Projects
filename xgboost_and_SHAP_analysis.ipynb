```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
data = pd.read_csv('/content/marketing_campaign (1).csv', delimiter='\t')

```

```python
import xgboost
import shap

# train an XGBoost model
X=data[['AcceptedCmp1','AcceptedCmp2','AcceptedCmp3','AcceptedCmp4','AcceptedCmp5','Response']]
y=data['NumDealsPurchases']
model = xgboost.XGBRegressor().fit(X, y)

# model's predictions using SHAP

explainer = shap.Explainer(model)
shap_values = explainer(X)

# visualize the first prediction's explanation
shap.plots.waterfall(shap_values[0])
```

```python
# visualize the first prediction's explanation with a force plot
shap.initjs()
shap.plots.force(shap_values[0])
```

```python
# visualize all the training set predictions
shap.initjs()
shap.plots.force(shap_values)
```

```python
# summarize the effects of all the features
shap.initjs()
shap.plots.beeswarm(shap_values)
```

```python
shap.initjs()
shap.plots.bar(shap_values)
```

```python
shap.initjs()
# plot the SHAP values for the Setosa output of all instances
shap.TreeExplainer(model).shap_interaction_values(X)
```
